```python
#名前(name)と年齢(age)の属性を持つHumanクラスを作成してください。
#nameとageを標準出力(print)するメソッド(printinfo)を追加してください。
#Humanクラスのインスタンスは、変数に代入してプログラム内で使用してください。

class Human:
  def __init__(self,name,age):
    self.name = name
    self.age = age
  def printinfo(self):
    print(self.name)
    print(self.age)

Nagisa = Human("なぎさ",18)
Akari = Human("あかり",6)

print(f"名前は{Nagisa.name}です")
print(f"年齢は{Nagisa.age}歳です")

print(f"名前は{Akari.name}です")
print(f"年齢は{Akari.age}歳です")
```
