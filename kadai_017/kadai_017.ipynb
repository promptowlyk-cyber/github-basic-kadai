```python
class Human :
  def __init__(self,name,age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20 :
      print(f"{self.name}は大人です")
    else :
      print(f"{self.name}は大人ではありません")

person = [
    Human("鈴木太郎", 37),
    Human("佐藤ひなた" , 12),
    Human("渡辺瑠璃" , 25),
    Human( "染谷淳" , 10),
    Human("木梨久留巳" , 18),
    Human("曽ヶ端明美" , 13),
    Human("佐伯亮" , 31),
    Human("工藤純一" , 52),
    Human("阿部まりや" , 46),
    Human("油井保奈美" , 34)
]

for x in person:
    x.check_adult()
```
